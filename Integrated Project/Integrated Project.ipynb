```python
# Cell 2: Utility helper functions (general-purpose)
# --------------------------------------------------
# Purpose:
# - Provide small utility helpers used across the notebook
# - Each function resides in its own cell as requested

try:
    def safe_print_df_info(df: pd.DataFrame, name: str) -> None:
        """
        Print basic information about a DataFrame safely.
        """
        try:
            print(f"\n--- {name} ---")
            print(f"Shape: {df.shape}")
            print("Columns:", list(df.columns)[:10], "... (showing first 10)")
            print("Index dtype:", df.index.dtype)
            print("Head:")
            print(df.head(3))
        except Exception as inner_e:
            print(f"[ERROR] Failed to print info for {name}: {inner_e}")

    print("Utility helpers ready.")

except Exception as e:
    print(f"[ERROR] Defining utility helpers failed: {e}")
```

```python
# Cell 3: Load datasets and set datetime index
# --------------------------------------------
# Purpose:
# - Read train, test, and full datasets
# - Parse 'date' column to datetime and set as index
# - Keep all data in memory without saving to files

try:
    def load_datasets(
        train_path: str, test_path: str, full_path: str
    ) -> Tuple[pd.DataFrame, pd.DataFrame, pd.DataFrame]:
        """
        Load train, test, and full datasets and set 'date' as datetime index if present.
        """
        # Read CSVs
        train = pd.read_csv(train_path)
        test = pd.read_csv(test_path)
        full = pd.read_csv(full_path)

        # Ensure 'date' exists and set as index
        for df, name in [(train, "train"), (test, "test"), (full, "full")]:
            if "date" in df.columns:
                df["date"] = pd.to_datetime(df["date"], errors="coerce")
                df.set_index("date", inplace=True)
            else:
                print(f"[WARN] 'date' column not found in {name} dataset; index left unchanged.")

        return train, test, full

    train_df, test_df, full_df = load_datasets(DATA_PATH_TRAIN, DATA_PATH_TEST, DATA_PATH_FULL)
    safe_print_df_info(train_df, "Train dataset")
    safe_print_df_info(test_df, "Test dataset")
    safe_print_df_info(full_df, "Full dataset")
    print("Datasets loaded successfully.")

except FileNotFoundError as e:
    print(f"[ERROR] Dataset file not found: {e}")
except pd.errors.ParserError as e:
    print(f"[ERROR] CSV parsing error: {e}")
except Exception as e:
    print(f"[ERROR] Failed to load datasets: {e}")
```

```python
# Cell 4: Recovery Calculation Helper Function
# --------------------------------------------
# Purpose:
# - Provide a reusable function to compute metal recovery using the project formula
# - Handle division-by-zero, invalid values, and infinities safely
# - Return a clean pd.Series suitable for verification, diagnostics, or feature engineering
# - Keep logic modular so later cells can call this function without duplication

try:
    def compute_recovery(feed_col: str, conc_col: str, tail_col: str, df: pd.DataFrame) -> pd.Series:
        """
        Compute recovery using the formula:
        recovery = (C * (F - T)) / (F * (C - T))
        Handles division by zero and invalid values gracefully.
        """
        F = df[feed_col].astype(float)
        C = df[conc_col].astype(float)
        T = df[tail_col].astype(float)

        numerator = C * (F - T)
        denominator = F * (C - T)

        with np.errstate(divide="ignore", invalid="ignore"):
            rec = np.where(denominator != 0, numerator / denominator, np.nan)

        # Replace infinities with NaN
        rec = pd.Series(rec, index=df.index, name="recovery_calc").replace([np.inf, -np.inf], np.nan)
        return rec

except Exception as e:
    print(f"[ERROR] Recovery verification failed: {e}")
```

```python
# Cell 7: Analyze Au/Ag/Pb concentrations across purification stages
# ------------------------------------------------------------------
# Purpose:
# - Examine how metal concentrations change across stages
# - Print summary statistics by stage

try:
    def summarize_stage_concentrations(df: pd.DataFrame, metals=("au", "ag", "pb")) -> pd.DataFrame:
        """
        Summarize median concentrations for each metal across stages
        (rougher.output.concentrate_*, primary_cleaner.output.concentrate_*, secondary_cleaner.output.concentrate_*, final.output.*)
        """
        rows = []
        stages = ["rougher", "primary_cleaner", "secondary_cleaner", "final"]
        for stage in stages:
            for metal in metals:
                # Heuristic: columns containing stage + ('concentrate' or 'output') + metal
                candidates = [c for c in df.columns if stage in c and (("concentrate" in c) or ("output" in c)) and metal in c]
                if candidates:
                    median_val = df[candidates].median(numeric_only=True).mean()
                    rows.append({"stage": stage, "metal": metal.upper(), "median_concentration": float(median_val)})
        return pd.DataFrame(rows)

    stage_summary = summarize_stage_concentrations(train_df)
    if stage_summary.empty:
        print("[WARN] No stage concentration columns found with the expected naming pattern.")
    else:
        print("Median concentrations by stage (train):")
        print(stage_summary.pivot(index="stage", columns="metal", values="median_concentration").round(4))

except Exception as e:
    print(f"[ERROR] Stage concentration analysis failed: {e}")
```

```python
# Cell 9: Check total substance concentrations and handle anomalies (train only)
# ---------------------------------------------------------------------------------------
# Purpose:
# - Compute total concentrations (Au+Ag+Pb or available constituents) at key stages
# - Identify abnormal totals using quantile-based thresholds (e.g., 99th percentile)
# - Cap extreme values (winsorization) instead of excluding rows
# - Preserve all training rows (no deletion), consistent with project rules

try:
    def total_concentration(df: pd.DataFrame, stage_prefix: str, metals=("au", "ag", "pb")) -> pd.Series:
        """
        Sum concentrations for provided metals at a given stage.
        Uses heuristic column matching; returns NaN when missing.
        """
        cols = []
        for metal in metals:
            candidates = [c for c in df.columns if stage_prefix in c and metal in c]
            concentrate_candidates = [c for c in candidates if "concentrate" in c or "output" in c]
            if concentrate_candidates:
                cols.extend(concentrate_candidates)
            else:
                cols.extend(candidates)
        if not cols:
            return pd.Series(np.nan, index=df.index)
        return df[cols].sum(axis=1, numeric_only=True)

    # Compute totals (heuristic across stages)
    train_totals = pd.DataFrame({
        "raw_feed_total": total_concentration(train_df, "feed"),
        "rougher_total": total_concentration(train_df, "rougher"),
        "final_total": total_concentration(train_df, "final"),
    })

    # Winsorize totals at 99th percentile (cap extreme values instead of excluding rows)
    for col in train_totals.columns:
        cap = train_totals[col].quantile(0.99)
        train_totals[col] = np.where(train_totals[col] > cap, cap, train_totals[col])
        print(f"{col}: capped values above 99th percentile ({cap:.2f})")

    # Replace original totals in training view with capped values
    train_df["raw_feed_total"] = train_totals["raw_feed_total"]
    train_df["rougher_total"] = train_totals["rougher_total"]
    train_df["final_total"] = train_totals["final_total"]

    # Keep all rows for training (no deletion)
    X_train_filtered = train_df.drop(columns=TARGETS).copy()
    y_train_filtered = train_df[TARGETS].copy()
    print(f"Training rows retained after winsorization: {X_train_filtered.shape[0]} of {train_df.shape[0]}")

except Exception as e:
    print(f"[ERROR] Total concentration anomaly handling failed: {e}")
```

```python
# Cell 9.1: Detect and Remove Near-Zero Total Concentrations (Train + Test)
# -----------------------------------------------------------------------
# Purpose:
# - Reuse the total_concentration() helper from Cell 9
# - Compute total concentrations for train and test
# - Clean inf/NaN/non-positive values before plotting
# - Plot distributions safely (linear + log)
# - Identify and remove near-zero total concentration rows

try:
    # 1. Compute total concentrations using the function from Cell 9
    train_totals = pd.DataFrame({
        "raw_feed_total": total_concentration(train_df, "feed"),
        "rougher_total": total_concentration(train_df, "rougher"),
        "final_total": total_concentration(train_df, "final"),
    })

    test_totals = pd.DataFrame({
        "raw_feed_total": total_concentration(test_df, "feed"),
        "rougher_total": total_concentration(test_df, "rougher"),
        "final_total": total_concentration(test_df, "final"),
    })

    # 2. Clean totals BEFORE plotting
    def clean_for_plot(series):
        s = series.replace([np.inf, -np.inf], np.nan)
        s = s.dropna()
        s = s[s > 0]   # log scale cannot handle zero or negative values
        return s

    train_totals_clean = train_totals.apply(clean_for_plot)

    # 3. Plot distributions safely
    fig, axes = plt.subplots(3, 2, figsize=(14, 12))
    stages = ["raw_feed_total", "rougher_total", "final_total"]

    for i, stage in enumerate(stages):
        # Linear scale
        sns.histplot(train_totals_clean[stage], bins=100, ax=axes[i, 0], color="steelblue")
        axes[i, 0].set_title(f"Train {stage} (Linear Scale)")

        # Log scale
        sns.histplot(train_totals_clean[stage], bins=100, ax=axes[i, 1], color="darkred", log_scale=True)
        axes[i, 1].set_title(f"Train {stage} (Log Scale)")

    plt.tight_layout()
    plt.show()

    # 4. Define near-zero threshold
    threshold = 1e-6

    # 5. Identify near-zero rows
    train_zero_mask = (
        (train_totals["raw_feed_total"] <= threshold) |
        (train_totals["rougher_total"] <= threshold) |
        (train_totals["final_total"] <= threshold)
    )

    test_zero_mask = (
        (test_totals["raw_feed_total"] <= threshold) |
        (test_totals["rougher_total"] <= threshold) |
        (test_totals["final_total"] <= threshold)
    )

    # 6. Remove near-zero rows
    train_removed = train_zero_mask.sum()
    test_removed = test_zero_mask.sum()

    train_df = train_df.loc[~train_zero_mask].copy()
    test_df = test_df.loc[~test_zero_mask].copy()

    print("\n[Near-Zero Total Concentration Filtering]")
    print(f"- Threshold used: {threshold}")
    print(f"- Train rows removed: {train_removed}")
    print(f"- Test rows removed: {test_removed}")
    print(f"- Train remaining: {train_df.shape[0]}")
    print(f"- Test remaining: {test_df.shape[0]}")

except Exception as e:
    print(f"[ERROR] Near-zero detection failed: {e}")
```

```python
# Cell 10: sMAPE function
# ---------------------------------------

def smape(y_true: np.ndarray, y_pred: np.ndarray) -> float:
    """
    Compute sMAPE (%) using the project formula:
    sMAPE = (1/N) * Œ£ |y - y_hat| / ((|y| + |y_hat|) / 2) * 100
    """
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)

    denom = (np.abs(y_true) + np.abs(y_pred)) / 2.0
    with np.errstate(divide="ignore", invalid="ignore"):
        frac = np.where(denom != 0, np.abs(y_true - y_pred) / denom, 0.0)

    return float(np.nanmean(frac) * 100.0)

print("sMAPE function ready.")
```

```python
# Cell 11: Final weighted sMAPE function
# -------------------------------------
# Purpose:
# - Implement the weighted final sMAPE:
#   Final sMAPE = 25% * sMAPE(rougher) + 75% * sMAPE(final)

try:
    def final_weighted_smape(rougher_true: np.ndarray, rougher_pred: np.ndarray,
                             final_true: np.ndarray, final_pred: np.ndarray) -> float:
        """
        Compute weighted final sMAPE: 0.25 * sMAPE(rougher) + 0.75 * sMAPE(final)
        """
        s_r = smape(rougher_true, rougher_pred)
        s_f = smape(final_true, final_pred)
        return 0.25 * s_r + 0.75 * s_f

    print("Final weighted sMAPE function ready.")

except Exception as e:
    print(f"[ERROR] Defining final weighted sMAPE failed: {e}")
```
